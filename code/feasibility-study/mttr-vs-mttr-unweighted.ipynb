```python
import pandas as pd

# Read the CSV files
mttr_df = pd.read_csv('../../data/mttr/mttr.csv')
mttr_weighted_df = pd.read_csv('../../data/mttr/mttr_weighted.csv')

# Inner merge based on system_name and from_package_name
merged_df = pd.merge(mttr_df, mttr_weighted_df, on=['system_name', 'from_package_name'], how='inner')

# Convert relevant columns to numeric, handling errors
merged_df['mttr_expo_weighted'] = pd.to_numeric(merged_df['mttr_expo_weighted'], errors='coerce')
merged_df['avg_post_fix_exposure_time'] = pd.to_numeric(merged_df['avg_post_fix_exposure_time'], errors='coerce')

# Calculate the difference and sort by highest difference
merged_df['exposure_diff'] = merged_df['avg_post_fix_exposure_time'] - merged_df['mttr_expo_weighted']
merged_df = merged_df.sort_values('exposure_diff', ascending=False)

merged_df.head(5)
```
